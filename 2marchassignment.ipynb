Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

A scatter plot is a type of graph that displays the values of two variables as points in a two-dimensional coordinate system. Each point on the plot represents a single data point, and the position of the point on the x and y axes indicates the value of the corresponding variable.

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is scatter plot")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function in matplotlib is used to create a grid of subplots within a single figure. This function takes three integer arguments: the number of rows of subplots, the number of columns of subplots, and the index of the current subplot.



```python
plt.plot(x1,y1,x2,y2,x3,y3,x4,y4)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
